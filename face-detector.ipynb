## Importing modules

```python
import cv2
import matplotlib.pyplot as plt


imagePath = 'sample.jpg'
```

### Loading and Reading an Image Using OpenCV

```python
img = cv2.imread(imagePath)
```

```python
img.shape
```

### Converting an RGB Image to Grayscale and Obtaining Its Dimensions

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
gray_image.shape
```

### Detecting Faces in a Grayscale Image Using a Haar Cascade Classifier

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
face = face_classifier.detectMultiScale(
    gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(40, 40)
)
```

```python
for (x, y, w, h) in face:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 4)
```

### Drawing Rectangles Around Detected Faces and Converting Image to RGB Format

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

### Plotting final image

```python
plt.figure(figsize=(20,10))
plt.imshow(img_rgb)
plt.axis('off')
```

### Initializing Face Detection with a Haar Cascade Classifier and Webcam Capture

```python
face_classifier = cv2.CascadeClassifier(
    cv2.data.haarcascades + "haarcascade_frontalface_default.xml"
)
```

```python
video_capture = cv2.VideoCapture(0)
```

### Detecting and Drawing Bounding Boxes Around Faces in a Video Frame

```python
def detect_bounding_box(vid):
    gray_image = cv2.cvtColor(vid, cv2.COLOR_BGR2GRAY)
    faces = face_classifier.detectMultiScale(gray_image, 1.1, 5, minSize=(40, 40))
    for (x, y, w, h) in faces:
        cv2.rectangle(vid, (x, y), (x + w, y + h), (0, 255, 0), 4)
    return faces
```

```python
while True:

    result, video_frame = video_capture.read()  
    if result is False:
        break  
    faces = detect_bounding_box(
        video_frame
    )  
    cv2.imshow(
        "My Face Detection Project", video_frame
    )  
    if cv2.waitKey(1) & 0xFF == ord("q"):
        break

video_capture.release()
cv2.destroyAllWindows()
```
